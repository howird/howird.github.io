```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
px = [3.5, 4]
py = [3.5, 4]

nx = [1, 2]
ny = [2, 1]

l1x = np.array([0, 5])
l1y = np.array([5, 0])

l2x = np.array([0, 6])
l2y = np.array([6, 0])

l3x = np.array([0, 4])
l3y = np.array([4, 0])
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(px, py, marker="+")
ax.scatter(nx, ny, marker="_")

ax.set_ylim([0, 6])
ax.set_xlim([0, 8])
ax.set_aspect('equal')
```

```python
fig.savefig('img/4.1.1.png')
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(px, py, marker="+")
ax.scatter(nx, ny, marker="_")
ax.plot(l1x, l1y, label = "decision boundary", color="tab:red", linestyle='dashed')
ax.plot(l3x, l3y, label = "margin", color="tab:green")
ax.plot(l2x, l2y, color="tab:green")

ax.set_xlim([0, 8])
ax.set_aspect('equal')

ax.legend()
```

```python
fig.savefig('img/4.1.2.png')
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(px, py, marker="+")
ax.scatter(nx, ny, marker="_")
ax.plot(l1x, l1y, color="tab:red", linestyle='dashed')
ax.plot(l2x, l2y, color="tab:green")
ax.plot(l3x, l3y, color="tab:green")

ax.set_xlim([0, 8])
ax.set_aspect('equal')

ax.quiver([0], [0], [1], [1], angles='xy', scale_units='xy', 
          scale=1, color="tab:olive", label=r"$\bar w$")
ax.quiver([0], [0], [1], [4.5], angles='xy', scale_units='xy', 
          scale=1, color="tab:purple", label=r"$\bar u$")
ax.legend()
```

```python
fig.savefig('img/4.1.3.png')
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(px, py, marker="+")
ax.scatter(nx, ny, marker="_")
ax.plot(l1x, l1y, color="tab:red", linestyle='dashed')
ax.plot(l2x, l2y, color="tab:green")
ax.plot(l3x, l3y, color="tab:green")

ax.set_xlim([0, 8])
ax.set_aspect('equal')

ax.quiver([0], [0], [1], [3], angles='xy', scale_units='xy', 
          scale=1, color="tab:purple", label=r"$x_-$")
ax.quiver([0], [0], [3], [3], angles='xy', scale_units='xy', 
          scale=1, color="tab:cyan", label=r"$x_+$")
ax.legend()
```

```python
fig.savefig('img/4.1.4.png')
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(px, py, marker="+")
ax.scatter(nx, ny, marker="_")
ax.plot(l1x, l1y, color="tab:red", linestyle='dashed')
ax.plot(l2x, l2y, color="tab:green")
ax.plot(l3x, l3y, color="tab:green")

ax.set_xlim([0, 8])
ax.set_aspect('equal')

ax.quiver([0], [0], [1], [3], angles='xy', scale_units='xy', 
          scale=1, color="tab:purple", label=r"$x_-$")
ax.quiver([0], [0], [3], [3], angles='xy', scale_units='xy', 
          scale=1, color="tab:cyan", label=r"$x_+$")

ax.quiver([1], [3], [2**-0.5], [2**-0.5], angles='xy', scale_units='xy', 
          scale=1, color="tab:olive", label=r"$\hat w$")

ax.quiver([1], [3], [2], [0], angles='xy', scale_units='xy', 
          scale=1, color="tab:pink", label=r"$x_+ - x_-$")

ax.legend()
```

```python
fig.savefig('img/4.1.4.png')
```
